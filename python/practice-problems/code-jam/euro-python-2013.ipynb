```python
#input = open('A-small-practice-1.in', 'r')
#output = open('A-small-practice-1.out', 'w')
input = open('A-small-practice-2.in', 'r')
output = open('A-small-practice-2.out', 'w')
num_test_cases = int(input.readline())
#num_test_cases = 10
for i in range(0, num_test_cases):
    num_skaters = int(input.readline())
    sorted_skaters = []
    cost = 0
    for j in range(0, num_skaters):
        skater_name = input.readline().strip()
        #print(skater_name)
        pos = len(sorted_skaters)
        for k in range(0, len(sorted_skaters)):
            if (skater_name < sorted_skaters[k]):
                pos = k
                cost += 1
                break;
        sorted_skaters.insert(pos, skater_name)
        #print(sorted_skaters)
    outputLine = 'Case #' + str(i + 1) + ': ' + str(cost) + '\n'
    #print('Case #' + str(i + 1) + ': ' + str(cost))
    print(outputLine)
    output.write(outputLine)
    
```
