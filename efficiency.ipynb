# Factorial goes beyond the scale
Must use logarithmic y axis scale to be able to show other time complexities on the same plot.

```python
x = np.arange(1, 11, 1)

factorial = np.vectorize(math.factorial)

plt.plot(x, factorial(x), label='n!')

plt.plot(x, 2**x, label='2^n')
plt.plot(x, x**2, label='n^2')
plt.plot(x, n * np.log(x), label='n log n')
plt.plot(x, x, label='n')
plt.plot(x, np.log(x), label='log n')
plt.plot(x,  np.array([1 for i in range(0, x.size)]), 'r', label='1')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.yscale('log', basey=10)
#plt.ylim([0, 40])
plt.xlim([2, 10])
plt.show()
```
